```python
findata.EPOCHS=200
pipeline.IS_VERBOSE = False
tickers = [
           'ABNB', 'ACLS', 'AFRM' ,'AI', 'AMBA', 'APO', 'BAX',
           'BILL', 'BMBL', 'CFLT', 'CRCT', 'CRWD',
           'DDOG', 'DOCN', 'DOCS', 'DOCU',
           'EXPE', 'FTCH', 'GLBE','HSII',
           'INMD', 'JWN', 'M', 'MDB',  'MNDY', 'MPWR',  'MQ', 'MXL',
           'MTCH', 'MYTE', 'NET', 'NVAX', 'NVCR',
           'OKTA', 'PAR', 'PTON', 'RBLX',  'RH', 'REGN', 'SNOW',
           'SFIX', 'STEM', 'SWAV',
           'TDOC', 'TOST','TREX', 'TWLO', 'TWTR', 
           'U','UBER','VRTX', 'Z', 'ZEN', 'ZM']
lossfn = "huber_loss"
mod = pipeline.RateReturnOnly(
    pipeline.FeatureSeq([pipeline.AddDayMonth(), pipeline.AddVWap(), pipeline.AddMA(200)]))
df  = pipeline.runModelCombined(tickers, 'ipos2b', mod, True, loss=lossfn)[0]
```

```python
df.sort_values('Gain', ascending=False)
```
